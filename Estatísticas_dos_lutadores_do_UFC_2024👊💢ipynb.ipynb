#importação

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/content/ufc-fighters-statistics.csv')
```

# Manipulação

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ['nickname'])
```

```python
df.drop(columns = ['reach_in_cm'])
```

```python
df.drop(columns = ['nickname'])
```

```python
#nova variavel data de aniversario e coluna idade
data_de_aniversario = pd.to_datetime(df['date_of_birth'])
df['idade'] = 2024 - data_de_aniversario.dt.year
df['idade'].head(5)
```

```python
df['idade'].dropna()
```

```python
#criando nova coluna luta

df['lutas'] = df['wins'] + df['losses'] + df['draws']
df['lutas']
```

```python
#rate de lutas
df['win_rate'] = df['wins']/df['lutas']
df['win_rate']
```

# vizualização

idade dos lutadores

```python
plt.hist(df['idade'])
plt.title('media de idades')
plt.xlabel('idade')
plt.ylabel('quantidade')
plt.show()
```

os estilos mais efetivos

```python
#importante anotar
graf = df.groupby('stance').agg({'win_rate':'mean'}).reset_index()
plt.hist(x = graf)
plt.title('Estilos de lutas mais efetivos')
plt.show()
```

performance dos lutadores

```python
graf2 = df.groupby('name').agg({'win_rate':'mean'}).reset_index()
grafico_perfomance = graf2.head(10)
plt.hist(grafico_perfomance)
plt.xticks(rotation = 90)
plt.show()
plt.title('grafico de perfomance')
```

lutador que mais disfere golpes por minuto

```python
graf3 = df.groupby('name').agg({'significant_strikes_landed_per_minute':'mean'}).reset_index()
golpes = graf3.head(10)
plt.hist(x = golpes)
plt.title('golpes por minuto')
plt.xticks(rotation = 90)
plt.show()
```

porcentagem de ataques significativos que acertam com sucesso para o lutador.

```python
graf4 = df.groupby('name').agg({'significant_striking_accuracy':'mean'}).reset_index()
graf4.head(10)
```
