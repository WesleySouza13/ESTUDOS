```python
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/top_hiphop_artists_tracks.csv')
df
```

```python
#organizando dados
df.columns
graf1 = df.groupby('Artist').count()
graf1= df.groupby('Artist').count().Popularity.sort_values(ascending = False)
```

```python
#plotar grafico de popularidade por artistas
plt.figure(figsize = (20,7))
plt.bar(graf1.index,graf1.values)
plt.xticks(rotation = 90)
plt.show
```
